```python
lat = 41.7092
lon = -87.9820
max_lat = 39
min_lat =35
min_lon = -100
max_lon = -94.5
display = pyart.graph.RadarMapDisplay(myradar)
f = plt.figure(figsize = [18,10])
display.plot_ppi_map('reflectivity', max_lat = max_lat, min_lat =min_lat,
                     min_lon = min_lon, max_lon = max_lon, vmin = -8,
                     vmax = 64, lat_lines = np.arange(min_lat,max_lat,.5),
                     lon_lines = np.arange(min_lon, max_lon, 1),
                     resolution = 'l')
for i in range(len(mydict['KVNX']['labels']['text'])):
    display.plot_point(mydict['KVNX']['labels']['locations'][i][1],
                       mydict['KVNX']['labels']['locations'][i][0],
                       symbol='ro', label_text=mydict['KVNX']['labels']['text'][i], label_offset=(0, 0))
display.basemap.drawmapscale(-99.0, 38.7, -98, 37, 100.0, barstyle='fancy')
```
